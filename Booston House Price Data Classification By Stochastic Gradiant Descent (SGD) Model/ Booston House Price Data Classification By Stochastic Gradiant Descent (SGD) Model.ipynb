Scaling function use for a purpose to scale data in manner form in which all values come to the same scale,because of its working this algorithm react faster to help the model for the accurate result

```python
x_scalar=StandardScaler()
y_scalar=StandardScaler()
```

Here,we have to reshape the dependent data (y_train,y_test) in vector form for the standardscalar() function because it cannot scale the values in array form,so we have to change it and independent data (x_train,x_test) is already in vector form
